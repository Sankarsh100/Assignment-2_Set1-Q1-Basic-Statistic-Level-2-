```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
measure = pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
```

```python
name_company=['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert']
```

```python
#pie plot
fig = plt.figure(figsize=(15,8))
plt.pie(measure, labels = name_company, autopct = '%1.0f%%')
plt.title('Pie Chart for the Company Data')
plt.show
fig.savefig('Pie Chart for the Company Data',  dpi=300, bbox_inches='tight')
```

```python
#Outliers
sns.boxplot(measure)
```

From the above graph we can conclude “Morgan Stanley is an outlier of 
91.36”.
